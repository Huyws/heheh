```python
#Bằng các phương pháp hình thang và parabol
from sympy import*
f = lambda x: 1/((1+2*x)**2)
```

```python
import numpy as np
X = np.linspace(0, 1.5, 7)
Y = [ f(x) for x in X]
```

```python
X
```

```python
Y
```

```python
#Tính tích phân gần đúng hthang
sum ( [(X[i] - X[i-1]) * (Y[i] + Y[i-1])/2 for i in range (1,7)])
```
