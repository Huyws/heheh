```python
t = symbols('t')
plot(f(x),(x,-5,-2))
df = lambda x: f(t).diff().subs(t,x)
```

```python
#Xác định xấp xỉ ban đầu x0 và tìm nghiệm gần đúng tới x3.
#F''(x)>0 f''x>0 chon x0=3
#x(n+1)=xn-f(xn)/f'(xn)
x=-2
for _ in range (3) :
 x = x - f(x) / df(x)
 print(x)
```
