```python
#Lap diem bat dong
g= lambda x: (32-x)**(1/5)
```

```python
from sympy import*
x= symbols('x')
```

```python
plot(g(x), (x, 0, 10))
```

```python
plot(abs(g(x).diff()), (x, 0, 10))
q= 0.012
```

```python
x0= 5.
for _ in range (3):
    x = g(x)
    print(x)
```

```python
x0= 5.
for _ in range (3):
    x = g(x0)
    ss = q / (1 - q) * abs (x - x0)
    x0 = x
    print(x, ss)
```

```python
q = 0.38
x0 = 5.
n = 0

while True:
    x = g(x0)
    ss = q / (1-q)*abs(x-x0) 
    x0 = x 
    n += 1
    print(n, x, ss )
    if ss < 10**-4:
        break
```
